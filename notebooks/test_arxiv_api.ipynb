```python
import urllib, urllib.request
url = 'http://export.arxiv.org/api/query?search_query=all:electron&start=0&max_results=1'
url = 'http://export.arxiv.org/api/query?search_query=dirac'
data = urllib.request.urlopen(url)
xml_string = data.read().decode('utf-8')
```

```python
import xmltodict
```

```python
python_dict = xmltodict.parse(xml_string)
```

```python
import pandas as pd
```

```python
xml_string
```

```python
xml_dict = xmltodict.parse(xml_string)

entries = xml_dict['feed']['entry']
if not isinstance(entries, list):
    entries = [entries]

rows = []
for entry in entries:
    article_id = entry['id']
    title = entry['title']
    summary = entry['summary']
    published = entry['published']
    authors = entry['author']
    if not isinstance(authors, list):
        authors = [authors]
    for author_idx, author in enumerate(authors):
        name = author['name']
        affiliation = author.get('arxiv:affiliation', None)
        rows.append({
            'article_id': article_id,
            'author_idx': author_idx,
            'author': name,
            'affiliation': affiliation,
            'title': title,
            'summary': summary,
            'published': published
        })

df = pd.DataFrame(rows)
df.set_index(['article_id', 'author_idx'], inplace=True)
df
```
